```python
import os

SEED = 536

cmd = f"""
python /kaggle/working/visual_search/scripts/train_clip.py \
    --dataset_root "{DATASET_ROOT}" \
    --output_dir "{CKPT_DIR}/seed_{SEED}" \
    --epochs 8 \
    --batch_size 16 \
    --lr 2e-5 \
    --weight_decay 1e-2 \
    --unfreeze_last_n 6 \
    --embed_dim 768 \
    --triplet_margin 0.25 \
    --triplet_weight 0.35 \
    --temperature 0.07 \
    --seed {SEED} \
    --eval_every 1
"""

print(cmd)

exit_code = os.system(cmd)

if exit_code != 0:

    raise RuntimeError(
        f"Training failed for seed {SEED}"
    )
```
